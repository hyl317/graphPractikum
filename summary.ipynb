```python
import pickle
N = 10
for mode in ['default', 'clock', 'circ', 'circ_clock']:
    basepath = f'./simulations/N{N}/{mode}'
    outdir = f'{basepath}/results/WP3.2_3leaves'
    print(f'mode: {mode}')
    failed_index = pickle.load(open(f'{outdir}/failed_index', 'rb'))
    print(f'number of failed r-maps: {len(failed_index)}')
    common_triple = pickle.load(open(f'{outdir}/common_triple', 'rb'))
    print(f'average common triples: {np.mean(common_triple):.3f}')
    for i in range(0, N-3):
        print(f'{i} common triples: {np.sum(np.array(common_triple)==i)}')
    init_leaves_recovered = pickle.load(open(f'{outdir}/init_leaves_recovered', 'rb'))
    print(f'percent of recognized R-maps with the initial 4 leaves recovered: {np.mean(init_leaves_recovered):.3f}')
    runtime = pickle.load(open(f'{outdir}/runtime', 'rb'))
    print(f'average runtime: {round(np.mean(runtime), 3)}')

```

```python
import os
import pickle

tasks = ['WP3.1_3leaves', 'WP3.1_4leaves', 'WP3.2_3leaves', 'WP3.2_4leaves']
for task in tasks:
    for N in [6,8,10]:
        for mode in ['default', 'clock', 'circ', 'circ_clock']:
            if not os.path.isdir(f'./failed_Rmaps/WP3/{task}/N={N}/{mode}'):
                os.mkdir(f'./failed_Rmaps/WP3/{task}/N={N}/{mode}')
            dir2results = f'./simulations/N{N}/{mode}/results/{task}'
            if os.path.isdir(dir2results):
                failed_index = pickle.load(open(f'{dir2results}/failed_index', 'rb'))
                for r in failed_index:
                    os.system(f'cp ./simulations/N{N}/{mode}/hist_rep{r}.txt ./failed_Rmaps/{task}/N={N}/{mode}/')
```
